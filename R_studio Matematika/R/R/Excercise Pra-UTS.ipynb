```r
#SET

#1
Q <- c(-100:100)

#2
n <- readline()
n <- as.integer(n)
if(n %% 2 == 0){
    P <- seq(-100, 100, 2)
} else{
    P <- seq(-99, 99, 2)
}

#3
S <- c(1, 5, 0, 2, 8)
R <- setdiff(Q, S)

#4
"P union R"
union(P, R)

#5
"Q complement (P union R)"
setdiff(Q, union(P, R))

#6
"Q complement P union Q complement R"
intersect(setdiff(Q, P), setdiff(Q, R))
```

```r
#FUNCTION

#1
f <- function(x, y){
    return (sqrt(x) + y^2)
}
"No. 1"
f(6, 8)

#2
g <- function(a, b){
    return (a*b*(a^2 + b/3))
}
"No. 2"
g(6, 8)

#3
h <- function(x, y){
    return (sqrt(f(x, y) + 3 + g(x, y)))
}
"No. 3"
h(6, 8)

#4
f1f2 <- function(x){
    f1 <- function(x){
        return (x^3 + x + 1)
    }
    f2 <- function(x){
        return (sqrt(x) - 1)
    }
    
    return (f1(f2(x)))
}
"No. 4"
f1f2(16)

#5
"No. 5"
f1 <- function(x){
    return (1/x)
}
f2 <- function(x){
    return (2/x)
}
f3 <- function(x){
    return (3/x)
}
f4 <- function(x){
    return (4/x)
}
f5 <- function(x){
    return (5/x)
}

x <- c(-16:16)

plot(x, f1(x), ylab="y", type="l", col="red")
lines(x, f2(x), col="orange")
lines(x, f3(x), col="yellow")
lines(x, f4(x), col="green")
lines(x, f5(x), col="purple")
```

```r
library(Ryacas)
x <- Sym("x")
h <- Sym("h")
t <- Sym("t")

#LIMITS

#1
Limit((1 - cos(x))/x, x, 0)

#2
Limit((2*(-3 + h)^2 - 18)/h, h, 0)

#3
Limit((t - sqrt(3*t + 4))/(4 - t), t, 4)
```

```r
#DIFFERENTIATION

#1
"No. 1"
deriv(sqrt(x)*(x + 1), x)

#2
"No. 2"
deriv((2*x^2 - 3)/sqrt(x), x)

#3
"No. 3"
deriv((x - 1)/(x + 1), x)

#4
"No. 4.1"
d1 <- function(x){
  return(sqrt(x) + (x + 1)/(2*sqrt(x)))
}

"No. 4.2"
d2 <- function(x){
  return((sqrt(x) * (4 * x) - (2 * x^2 - 3)/(2 * sqrt(x)))/x)
}

"No. 4.3"
d3 <- function(x){
  return((x - (x - 1) + 1)/(x + 1)^2)
}
```
