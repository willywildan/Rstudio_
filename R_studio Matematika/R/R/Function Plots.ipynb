```r
#Constant functions

f <- function(x){
    #suppose c = 16
    fx <- 16
    return (fx)
}

input <- 0:22
plot(main = expression("f(x) = c"),
    input,
    sapply(input, f),
    type = "l",
    xlab = "x",
    ylab = "f(x)")
```

```r
#Linear functions

f <- function(x){
    #a = 6 and b = 8
    fx <- 6 * x + 8
    return (fx)
}

input <- -6:22
plot(main = expression("f(x) = ax + b"),
    input,
    sapply(input, f),
    type = "l",
    xlab = "x",
    ylab = "f(x)")
```

```r
#Quadratic functions

f <- function(x){
    #a = 6, b = 8, c = 16
    fx <- 6*x^2 + 8*x + 16
    return (fx)
}

input <- -22:22
plot(main = expression("f(x) = ax"^2*"+ bx + c"),
    input,
    sapply(input, f),
    type = "l",
    xlab = "x",
    ylab = "f(x)")
```

```r
#Polynomial functions

f <- function(x){
    #a = 1, b = 8, c = 6, d = 16
    fx <- x^3 - 8*x^2 + 6*x - 16
    return (fx)
}

input <- seq(-8, 16, 0.1)
plot(main = expression("f(x) = a"[n]*"x"^n*"+ a"[n-1]*
                       "x"^n-1*"+ ... + a"[1]*"x + a"[0]),
    input,
    sapply(input, f),
    type = "l",
    xlab = "x",
    ylab = "f(x)")
```

```r
#Rational functions

f <- function(x){
    fx <- (x^2 - x + 2)/(x - 3)
    return (fx)
}

input <- seq(6, 8, 0.16)
plot(main = expression("f(x) = p(x)/q(x)"),
    input,
    sapply(input, f),
    type = "l",
    xlab = "x",
    ylab = "f(x)")
```
